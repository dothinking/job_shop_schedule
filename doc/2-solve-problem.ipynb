# Solve Problem

Load a benchmark problem and solve it with built-in solver, e.g., `GoogleORCPSolver`

```python
from jsp import JSProblem
from jsp.solver import GoogleORCPSolver

# load benchmark problem
problem = JSProblem(benchmark='ft06')

# initialize solver
s = GoogleORCPSolver()
```

Call `JSSolver.solve(problem, interval, callback)` to solve the problem.
- `interval` A dynamic Gantt chart is shown during the solving process when `interval` is not None.
- `callback` Call user defined function once a better solution is found.

```python
s.solve(problem=problem,
    interval=2000,
    callback=lambda solution: print('current makespan: ', solution.makespan))

# NOTE: the solving process is running in a child-thread.
# wait and show result
s.wait()
print()
print('running:', s.is_running)
print('Solving time ', s.user_time)
print('Makespan ', s.solution.makespan)
```

```python
# explore solution
solution = s.solution
print('feasible solution:', solution.is_feasible)
for op in solution.ops:
    print('job=', op.source.job.id, 'machine=', op.source.machine.id, 'start_time=', op.start_time)
```

```python
# explore solution: Gantt chart
solution.plot_gantt_chart()
```
